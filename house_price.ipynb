```python
def compute_cost(x,y,w,b):
    m=x.shape[0]
    
    total_cost,i=0,0
    for i in range(m):
        f_wb=w*x+b
        cost=(f_wb-y[i])**2
        total_cost+=cost
        
    total_cost/=(2*m)
    
    return total_cost
```

```python
initial_w,initial_b=5,7

cost=compute_cost(x_train,y_train,initial_w,initial_b)
print(f"Cost at initial w,b:{round(cost,3)}")
```
